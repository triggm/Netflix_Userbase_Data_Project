# Introduction: Netflix Userbase Dataset
* This Seaborn notebook offers beginner-friendly techniques for exploratory data analysis (EDA) and visualization using the Netflix Userbase Dataset. With Seaborn's capabilities, I will analyze the data and extract valuable insights on gender distribution, age demographics, subscription types, and more. This notebook exemplifies my competence in data analysis and visualization, providing a valuable resource for beginners to develop hands-on experience with Seaborn, a renowned Python library for data exploration.

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Netflix userbase dataset
data = pd.read_csv('/kaggle/input/netflix-userbase-dataset/Netflix Userbase.csv')

# Exploratory Data Analysis (EDA)
# Let's start by inspecting the first few rows of the dataset
data.head()
```

# Let's See the Statistics

```python
# Basic statistics summary
data.describe()
```

# 1. Bar plot - Gender distributionÂ¶ 

```python
# Data visualization
# 1. Bar plot - Gender distribution
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# 2. Bar plot - Age group distribution


```python
# 2. Bar plot - Age group distribution
sns.countplot(x='Age', data=data)
plt.title('Age Group Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# 3. Pie chart - Subscription types

```python
# 3. Pie chart - Subscription types
subscription_counts = data['Subscription Type'].value_counts()
plt.pie(subscription_counts, labels=subscription_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Subscription Types')
plt.show()
```

# 4. Box plot - Monthly Revenue by Subscription 

```python
# 4. Box plot - Monthly revenue by subscription type
sns.boxplot(x='Subscription Type', y='Monthly Revenue', data=data)
plt.title('Monthly Revenue by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Monthly Revenue')
plt.show()
```

# 5. Heatmap (Correlation Matrix)

```python
# 5. Heatmap (correlation matrix)
corr_matrix = data[['Monthly Revenue', 'Age']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# 6. Bar plot - Country Distribution

```python
# 6. Bar plot - Country distribution
sns.countplot(y='Country', data=data)
plt.title('Country Distribution')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

# 7. Box plot - Monthly Revenue by Gender

```python
# 7. Box plot - Monthly revenue by gender
sns.boxplot(x='Gender', y='Monthly Revenue', data=data)
plt.title('Monthly Revenue by Gender')
plt.xlabel('Gender')
plt.ylabel('Monthly Revenue')
plt.show()
```

# 8. Violin plot - Age Distribution by Subscription 

```python
# 8. Violin plot - Age distribution by subscription type
sns.violinplot(x='Subscription Type', y='Age', data=data)
plt.title('Age Distribution by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Age')
plt.show()
```

# 9. Count plot - Device usage

```python
# 9. Count plot - Device usage
sns.countplot(x='Device', data=data)
plt.title('Device Usage')
plt.xlabel('Device')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

# 10. Scatter plot - Monthly Revenue vs. Plan Duration

```python
# 10. Scatter plot - Monthly revenue vs. Plan Duration
sns.scatterplot(x='Plan Duration', y='Monthly Revenue', data=data)
plt.title('Monthly Revenue vs. Plan Duration')
plt.xlabel('Plan Duration')
plt.ylabel('Monthly Revenue')
plt.show()
```
